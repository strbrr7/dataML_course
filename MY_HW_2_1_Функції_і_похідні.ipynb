```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = np.linspace(0, 10, 20)
f = 10 * x - 5
df = np.full_like(x, 10, float) # f`(x) = 10
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x, f, label = 'f(x)=10x-5', color = 'blue')
plt.plot(x, df, label='f`(x)=10')
```

Похідна функції дорівнює 10, що підтверджує сталу швидкість прогресу: кожна витрачена година навчання забезпечує рівномірний приріст оцінки на 10 балів, незалежно від того, на якому етапі навчання перебуває студент. Ефективність навчання не падає і не зростає з часомю

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f_sin = sp.sin(x)
derivative_sin = sp.diff(f_sin, x)

f_ln = sp.ln(x)
derivative_ln = sp.diff(f_ln, x)

print(f'Похідна sin(x): {derivative_sin}')
print(f'Похідна ln(x): {derivative_ln}')
```

```python
x_vals = np.linspace(1, 10)

ln_vals = np.log(x_vals)
ln_vals_derivative = 1 / x_vals

sin_vals = np.sin(x_vals)
sin_vals_derivative = np.cos(x_vals)

plt.plot(x_vals, ln_vals, label='ln(x)')
plt.plot(x_vals, ln_vals_derivative, label='(ln(x))`')
plt.plot(x_vals, sin_vals, label='sin(x)')
plt.plot(x_vals, sin_vals_derivative, label='(sin(x))`')

plt.legend()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
f_derivative = sp.diff(f, x)
```

```python
f_vals = sp.lambdify(x, f, 'numpy')
f_derivative_vals = sp.lambdify(x, f_derivative, 'numpy')
```

```python
type(f_vals)
```

```python
x_vals = np.linspace(0, 100)
```

```python
f_vals = f_vals(x_vals)
f_derivative_vals = f_derivative_vals(x_vals)
```

```python

plt.plot(x_vals, f_vals, label ='y=x^2-5x+1')
plt.plot(x_vals, f_derivative_vals, label='y`=2x-5')
plt.legend()
```
