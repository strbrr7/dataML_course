  2.1. без використання жодних бібліотек, користуючись тільки Python.  
  2.2. використовуючи відповідний метод бібліотеки `numpy`.

  Виведіть результат обчислень в кожному з випадків на екран та порівняйте результати. Вони співпадають?

```python
sum_purchases = [ 316,  493,  367,  153,  224,  121,  322,  450,  196,  397,  281,
  239,  484,  124,  223,  448,  432,  255, 4310, 3524]
```

```python
sample_mean = (1 / len(sum_purchases)) * sum(sum_purchases)
print(f'Без використання бібліотек: {sample_mean}')

sample_mean_np = np.mean(sum_purchases)
print(f'З використанням бібліотеки: {sample_mean_np}')
```

```python
median_no_np = (sorted(sum_purchases)[len(sum_purchases) // 2] + sorted(sum_purchases)[len(sum_purchases) // 2 -1]) / 2
print(f'Без викорстання бібліотеки: {median_no_np}')

median_np = np.median(sum_purchases)
print(f'З використанням бібліотеки: {median_np}')
```

**Завдання 4**. Обчислість стандартне відхилення наведеної вибірки в змінній `sum_purchases` кожним з зазначених методів:  

  4.1. без використання жодних бібліотек, користуючись тільки Python.

Виведіть результат обчислень в кожному з випадків на екран.

```python
std_no_np = np.sqrt(1 / (len(sum_purchases) - 1) * sum((el - sample_mean_np)**2 for el in sum_purchases))
print(f'Стандартне відхилення без використання бібліотеки: {std_no_np}')

std_np_sum_purchases = np.std(sum_purchases, ddof = 1)
print(f'Стандартне відхилення з використання бібліотеки: {std_np_sum_purchases}')
```

**Завдання 6**. Крім даних про суму покупок, ми зібрали ще дані про кількість відвідувань кожного з кожного покупця, які знаходяться в змінній `visits`. Обчисліть коефіцієнт кореляції Пірсона для суми покупок і кількості відвідувань кожним з зазначених методів:  

  6.1. без використання жодних бібліотек, користуючись тільки Python.  
  6.2. використовуючи відповідний метод бібліотеки numpy.  
  
Виведіть результат на екран. Проаналізуйте результат.

Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
mean_sum_purchases = sample_mean
mean_visits = sum(visits) / len(visits)

omega_sp = np.sqrt(sum(((x - mean_sum_purchases)**2 for x in sum_purchases)) / (len(sum_purchases) - 1))
omega_visits = np.sqrt((sum((y - mean_visits)**2 for y in visits)) / (len(visits) - 1))

xy = [(sum_purchases[i] - mean_sum_purchases) * (visits[i] - mean_visits)for i in range(len(sum_purchases))]
cov = sum(xy) / (len(visits) - 1)
corrcoef_no_np = cov / (omega_sp * omega_visits)
print(corrcoef_no_np)

corrcoef_np = np.corrcoef(sum_purchases, visits)
print(corrcoef_np)
```

**Висновок:** Обчислення кореляції з бібліотекою та без збігаються. Також кореляція слабка та негативна. Можна зробити висновки, що усі покупці робили покупки на одну середню суму, але є поодинокі випадки, коли покупці витрачали набагато більше (такби мовити викиди). Можна предположить, що ці дві змінні ніяк не впливають одне на одну. Висновок, якщо прибрати викиди, то можна стверджувати, що кількість візитів ніяк не впливає на суму покупки. Тобто кореляція буде близька до 0.

```python
import matplotlib.pyplot as plt

plt.scatter(sum_purchases, visits)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python

threshold = mean_sum_purchases + 2 * std_np_sum_purchases
clean_sum = []
clean_visits = []

for i in range(len(sum_purchases)):
  if sum_purchases[i] <= threshold:
    clean_visits.append(visits[i])
    clean_sum.append(sum_purchases[i])

clean_corrcoef = np.corrcoef(clean_sum, clean_visits)
print(clean_corrcoef)
```

```python
import matplotlib.pyplot as plt

plt.scatter(clean_sum, clean_visits)
plt.title('Коефіцієнт кореляції (без викидів)')
plt.xlabel('Sum of purchases')
plt.ylabel('Visits')
```

**Висновок:** Після усунення викидів коефіцієнт посильнішав. Тобто навіть після усунення викидів, ми можемо побачити, що чим більше користувач заходив в магазин, тим менше він купував. І навпаки чим менше користувач заходив, тим більше він купував.
