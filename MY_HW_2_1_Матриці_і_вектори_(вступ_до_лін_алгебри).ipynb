3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])
A = A.T

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])
B = B.T

C = np.array([[0],
              [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
result_matrix = np.dot(A, B)
```

```python
result_matrix, result_matrix.shape
```

Якщо ми домножимо `result_matrix`розмірністю 2х2 на `С` розмірністю 2х1, то вийде матриця розмірністю 2х1

```python
rm_mult_c = np.dot(result_matrix, C)
```

```python
rm_mult_c
```

Якщо ми домножимо result_matrixрозмірністю 2х1 на D розмірністю 2х2, то вийде матриця розмірністю 2х2

```python
rm_mult_d = np.dot(result_matrix, D)
```

```python
rm_mult_d
```
